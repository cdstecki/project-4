```python
import psycopg2
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv = 'Resources/data/data_cleaned_2021.csv'
```

```python
salary_df = pd.read_csv(csv).drop(['index', 'Salary Estimate', 'Job Description'], axis=1)
salary_df
```

```python
conn_string = 'postgres://user:stecki123@localhost/salary_db'
  
db = create_engine(conn_string)
conn = db.connect()

# our dataframe
data = salary_df
  
# Create DataFrame
df = pd.DataFrame(data)
df.to_sql('data', con=conn, if_exists='replace',
          index=False)
conn = psycopg2.connect(conn_string
                        )
conn.autocommit = True
cursor = conn.cursor()
  
sql1 = '''select * from data;'''
cursor.execute(sql1)
for i in cursor.fetchall():
    print(i)
  
# conn.commit()
conn.close()
```
