```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
salary_df = pd.read_csv("data_cleaned_2021.csv")
salary_df
```

```python
salary_df = salary_df.drop(columns = ['Salary Estimate','seniority_by_title','Age'])
salary_df
```

```python
salary_df["Avg Salary(K)"].unique()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import os
pd.set_option('display.max_rows', None)

job_sectors = salary_df ['Industry'].value_counts()
job_sectors = job_sectors[:13]
job_sectors

industry = pd.DataFrame(job_sectors)
#industry.drop(labels = ['-1'],axis=0, inplace=True)
industry.index.name = 'Industry Sectors'
industry

industry.plot.bar(figsize=(15,9), ylabel='Total Count',xlabel='Industry Type', title = 'Top Industries for Data Scientist Careers')
plt.show()

avg_sal = salary_df['Avg Salary(K)']

start=[]
end=[]
for x in avg_sal.tolist():
    try:
        val = x.split(' ')
        val = val[0].replace('K-',",")
        val = val.replace('K',"")
        val = val.split(",")
        end.append(float(val[1][1:]))
        start.append(float(val[0][1:]))
    except:
        end.append(1)
        start.append(1)

salary_df["new sal"] = [(start[x] + end[x]) / 2 for x in range(len(start))]
salary_df

sal_range = salary_df['new sal']
sal_range.describe()

title = salary_df['Job Title'].value_counts().sort_values(ascending=False)
title = title[:10]
title

job_title = pd.DataFrame(title)
job_title

job_title.plot.bar(figsize=(15,5), rot = 45, ylabel='Total Count', xlabel='Job Title', title = 'Top 10 Data Scientist Jobs')
plt.show()

city = salary_df['Location'].value_counts()
city = city[:10]
city

job_local = pd.DataFrame(city)
job_local

job_local.plot.pie(autopct='%1.1f%%',figsize=(8,8),y='Location',legend=None, title = "Top 10 Job Locations")
plt.show()
                        
```

```python
import pandas as pd
import os
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('display.max_rows', 500)

job_title_df= salary_df.loc[
    (salary_df["Job Description"].str.contains("Excel", case=False)) |
    (salary_df["Job Description"].str.contains("Python", case=False)) |
    (salary_df["Job Description"].str.contains("SQL", case=False)) |
    (salary_df["Job Description"].str.contains("Tableau", case=False)) |
    (salary_df["Job Description"].str.contains("Machine Learning", case=False))
]

jobs_langs = job_title_df

job_titles = salary_df["Job Title"].value_counts()
job_titles = job_titles[:20]
job_titles

keywords = [
    "Excel",
    "Python",
    "SQL",
    "Tableau",
    "Machine Learning"
]

x_axis = keywords
y_axis = []

for keyword in keywords:
    key_df = job_title_df.loc[job_title_df["Job Description"].str.contains(keyword, case=False)]
    num_valv = len(key_df)
    y_axis.append(num_valv)
    
plt.bar(x_axis, y_axis)
plt.xlabel("Top 5 Languages per Glassdoor")
plt.ylabel("Glassdoor Data Science Jobs 2021")
plt.title("Required Data Science Languages")

jc = plt.gca()
jc.tick_params(axis='x', colors='red')
jc.tick_params(axis='y', colors='green')

plt.show()
```

```python
jobs_langs.loc[jobs_langs["Job Description"].str.contains("Entry", case=False)]
jobs_langs.loc[jobs_langs["Job Description"].str.contains("Certification", case=False)]
jobs_langs.loc[jobs_langs["Job Description"].str.contains("Degree", case=False)]   

keywords = [
    "Degree",
    "Certification"
]

x_axis = keywords
y_axis = []

for keyword in keywords:
    key_df = job_title_df.loc[job_title_df["Job Description"].str.contains(keyword, case=False)]
    num_valv = len(key_df)
    y_axis.append(num_valv)
    
plt.bar(x_axis, y_axis)
plt.xlabel("Perferred Eduction per Job Descriptions")
plt.ylabel("Glassdoor Data Science Jobs 2021")
plt.title("Degree vs Certification")

jc = plt.gca()
jc.tick_params(axis='x', colors='red')
jc.tick_params(axis='y', colors='green')

plt.show()
```
