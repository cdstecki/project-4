```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
salary_df = pd.read_csv("data_cleaned_2021.csv")
salary_df
```

```python
city = salary_df['Location'].value_counts()
city = city[:10]
city
```

```python
job_local = pd.DataFrame(city)
job_local
```

```python
job_local.plot.bar(figsize=(15,5), rot = 45, ylabel='Salaries', xlabel='Top 10 Cities', title = 'Top 10 Average Salaries')
plt.show()
```
